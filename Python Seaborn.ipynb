```python
#working with more categorical data
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', col='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='sex')
```

```python
#facet grid
```

```python
FG = sns.FacetGrid(tips,row='day')
```

```python
FG = sns.FacetGrid(tips,col='day')
```

```python
FG = sns.FacetGrid(tips,col='day',row='sex')
```

```python
FG = sns.FacetGrid(tips,col='day',row='sex')
FG.map_dataframe(sns.histplot, x='tip')
```

```python
FG = sns.FacetGrid(tips,col='day',row='sex')
FG.map_dataframe(sns.scatterplot, x='tip', y='total_bill')
```
